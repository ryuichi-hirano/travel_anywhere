```python
import random
import pandas as pd
import numpy
import folium
```

```python
csvPath = '/content/drive/MyDrive/Archive/B1/Others/Colab_Notebooks_Others/Side_reaction/all_countries.csv'
```

```python
df = pd.read_csv(csvPath, header=None)
df
```

```python
add_df = df.drop([0], axis=1)
array_df = add_df.values.tolist()
```

```python
drop_col = ['jiscode', 'namekana', 'building', 'zipcode', 'tel', 'source', 'note']
add_df = df.drop(drop_col, axis=1)
array_df = add_df.values.tolist()
```

```python
random_row = random.randint(0, len(array_df))
goal_lat = [array_df[random_row][2], array_df[random_row][3]]
m = folium.Map(goal_lat, zoom_start=8)
folium.Marker(goal_lat,
              popup = array_df[random_row][0],
              tooltip = array_df[random_row][0]).add_to(m)
m.save("sheeePark.html")
m
```

```python
random_row = random.randint(0, len(array_df))
goal_lat = [array_df[random_row][0], array_df[random_row][1]]
m = folium.Map(goal_lat, zoom_start=8)
folium.Marker(goal_lat,
              popup = array_df[random_row][2],
              tooltip = array_df[random_row][2]).add_to(m)
m.save("sheeePark.html")
m
```
